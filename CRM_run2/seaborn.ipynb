```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.histplot(x='petal_length',data=iris)
```

```python
sns.histplot(x='petal_length',data=iris,bins=20,kde=True, color='red')
```

```python
sns.kdeplot(x='petal_length',data=iris,color='blue')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.kdeplot(x='total_bill',hue='time',fill=True,alpha=0.7,linewidth=2,data=tips)
```

```python
sns.kdeplot(x='total_bill',hue='day',fill=True,alpha=0.7,linewidth=2,data=tips)
```
