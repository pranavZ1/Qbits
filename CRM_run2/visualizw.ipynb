```python
import pandas as pd
```

```python
df = pd.read_csv("HistoricalPrices.csv")
```

```python
df
```

```python
df[' Open']
```

```python
df = df.rename(
    columns={
        ' Open':'Open',
        ' High':'High',
        ' Low':'Low',
        ' Close':'Close'
    }
)
```

```python
df['Open']
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by = 'Date')
```

```python
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Date'],df['Close'],label = 'close')
plt.plot(df['Date'],df['Open'],label='open')
plt.legend()
plt.show()
```

## Bar Chart 
- vertical
- horizontal

```python
from calendar import month_name
```

```python
df['Month'] = pd.Categorical(df['Date'].dt.month_name(),month_name[1:])
```

```python
df
```

```python
df["Month"].value_counts()
```

```python
dfMonth = df.groupby('Month').mean().reset_index()
```

```python
dfMonth
```

```python
plt.bar(dfMonth['Month'],height=dfMonth['Close'])
plt.xlabel('Months')
plt.ylabel('Closing Price')
plt.title('Bar Chart')
plt.show()
```

## Scatter plot

```python
plt.scatter(df['Open'],df['Close'])
plt.show()
```

## Trend Line:
- least square polynomial fit
- np.polyfit()
- np.poly1d()

```python
import numpy as np
```

```python
z = np.polyfit(df['Open'], df['Close'], 1)
p = np.poly1d(z)
```

```python
plt.scatter(df['Open'],df['Close'])
plt.plot(df['Open'],p(df['Open']),color = 'red')
plt.savefig("Scatter.png")
plt.show()
```
