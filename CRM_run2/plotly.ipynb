```python
import pandas as pd
```

```python
df = pd.read_csv('time.csv')
```

```python
import plotly.offline as py
import plotly.graph_objects as go
```

```python
df
```

```python
df.info()
```

```python
data = df.iloc[:100,:]
trace1 = go.Scatter(
    x=data.world_rank,
    y=data.citations,
    name='citations',
    text=data.institution,
    mode='lines',
)
trace2 = go.Scatter(
    x=data.world_rank,
    y=data.score,
    name='Score',
    text=data.institution,
    mode='lines',
)
layout = dict(
    title = 'Citation and score of World Rank of Top 100 Universities',
    xaxis = dict(title = 'World Rank',ticklen = 5,zeroline = False),
)

fig = dict(data = [trace1,trace2], layout = layout)
py.plot(fig)
```

```python
df2014 = df[df['year']==2014].iloc[:100,:]
df2015 = df[df['year']==2015].iloc[:100,:]
df2016 = df[df['year']==2016].iloc[:100,:]
```

```python
df2014
```

```python
trace1 = go.Scatter(
    x=df2014.world_rank,
    y=df2014.citations,
    name='2014',
    text=data.institution,
    mode='markers',
    marker=dict(color = 'rgba(16,112,2,0.8)'),
)
trace2 = go.Scatter(
    x=df2015.world_rank,
    y=df2015.citations,
    name='2015',
    text=data.institution,
    mode='markers',
    marker=dict(color = 'rgba(255,112,2,0.8)'),
)
trace3 = go.Scatter(
    x=df2016.world_rank,
    y=df2016.citations,
    name='2016',
    text=data.institution,
    mode='markers',
    marker=dict(color = 'rgba(0,128,2,0.8)'),
)

layout = dict(
    title = 'Citation vs world rank of top 100 universities with 2014,2015 and 2016',
    xaxis = dict(title = 'World Rank',ticklen = 5,zeroline = False),
    yaxis = dict(title = 'Citration',ticklen = 5,zeroline = False)
)

fig = dict(data = [trace1,trace2,trace3],layout=layout)
py.plot(fig)
```

```python
df2013 = df[df['year']==2013].iloc[:3,:]
```

```python
df2013
```

```python
trace1 = go.Bar(
    x = df2013.institution,
    y = df2013.citations,
    name='citations',
    marker=dict(color = 'rgba(255,174,255.0.5)',line = dict(color='rgb(0,0,0)',width = 1.5)),
    text='country',
)

trace2 = go.Bar(
    x = df2013.institution,
    y = df2013.score,
    name='Score',
    marker=dict(color = 'rgba(255,255,128.0.5)',line = dict(color='rgb(0,0,0)',width = 1.5)),
    text='country',
)

layout = go.Layout(barmode = 'group')
fig = go.Figure(data=[trace1,trace2],layout=layout)
py.plot(fig)
```
