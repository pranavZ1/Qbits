```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.offline as py
import plotly.graph_objects as go
```

```python
data = pd.read_csv('multipleChoiceResponses.csv',encoding='ISO-8859-1')
```

```python
data.head()
```

```python
df = pd.DataFrame(data['Country'].value_counts())
data
```

```python
df['country'] = df.index
```

```python
df.columns = ['number','country']
```

```python
data
```

```python
df = df.reset_index().drop('Country',axis=1)
```

```python
data
```

```python
d =[
    dict(
        type = 'chloropleth',
        locations = df['country'],
        locationmode = 'country names',
        z = df['number'],
        text = df['country'],
        colorscale = [[0,"rgb(5,10,172)"],[0.35,"rgb(40,60,190)"],[0.5,"rgb(70,100,245)"],[0.6,"rgb(90,120,245)"],[0.7,"rgb(106,137,247)"],[1,"rgb(220,220,220)"]],
        autocolor = False,
        reversescale = True,
        marker = dict(
            line = dict(
                color = 'rgb(180,180,180)',
                width = 0.5,
            )
        ),
        colorbar = dict(
            autotick = False,
            tickprefix = "$",
            title = "Number of Respondents",
        )
    )
]
```

```python
layout = dict(
    title = 'The nationality of Respondants',
    geo = dict(
        showframe = False,
        showcoastlines = True,
        projection = dict(
            type = "Mercator"
        )
    )
)
```

```python
fig = dict(data=d,layout=layout)
py.plot(fig, validate=False)
```
